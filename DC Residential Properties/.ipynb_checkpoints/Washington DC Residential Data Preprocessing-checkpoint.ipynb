# Washington DC Residential Analysis

Washington, D.C. is the capital of the United States. Washington's population is approaching 700,000 people, and has been growing since 2000 following a half-century of population decline. The city is highly segregated and features a high cost of living. In 2017, the average price of a single family home in the district was $649,000. This dataset provides insight on the housing stock of the district.

> **All data is available at Open Data D.C.. The residential and address point data is managed by the Office of the Chief Technology Officer**

**Explanation about this data**
<br>
There are variables not included for predict in this raw dataset. Here is an explanation about this data.
* **BATHRM** - Number of Full Bathrooms
* **HF_BATHRM** - Number of Half Bathrooms (no bathtub or shower)
* HEAT - Water Heater for bathrooms
* AC - Cooling rooms
* NUM_UNITS - Number of Units
* ROOMS - Number of Rooms
* BEDRM - Number of Bedrooms
* AYB - The earliest time the main portion of the building was built
* YR_RMDL - Year structure was remodeled
* EYB - The year an improvement was built more recent than actual year built
* STORIES - Number of stories in primary dwelling
* SALEDATE - Date of most recent sale
* PRICE - Price of most recent sale
* QUALIFIED - According to government staff, this is an internally-used indicator to reflect if a sale is representative of market value according to the office's internal criteria (for example, if it is part of a multi-parcel sale). U (Qualified) & Q (Qualified)
* SALE_NUM - Sale Number
* GBA - Gross building area in square feet
* BLDG_NUM - Building Number on Property
* STYLE - Style
* STRUCT - Structure
* GRADE - Grade
* CNDTN - Condition property
* EXTWALL - Extrerior wall
* ROOF - Roof type
* INTWALL - Interior wall
* KITCHEN - Number of kitchens
* FIREPLACES - Number of fireplaces
* LANDARE - Land area of property in square feet
* WARD - Ward (District is divided into eight wards, each with approximately 75,000 residents)

## Data Preprocessing 

**Kolom to drop** Unnamed: 0, CMPLX_NUM, LIVING_GBA , ASSESSMENT_SUBNBHD, CENSUS_TRACT, CENSUS_BLOCK, GIS_LAST_MOD_DTTM, STORIES, SALE_NUM, USECODE, CITY, STATE

```python
df.drop(['Unnamed: 0', "CMPLX_NUM", "LIVING_GBA" , "ASSESSMENT_SUBNBHD", "CENSUS_TRACT", 
         "CENSUS_BLOCK", "GIS_LAST_MOD_DTTM", "STORIES", "SALE_NUM", "USECODE", "CITY", 
         "STATE", "NATIONALGRID",'X','Y','SALEDATE','FULLADDRESS'],axis=1,inplace=True)
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total,percent], axis=1, keys = ['Total','Percent'])
missing_data
```

*Unnamed: 0, CMPLX_NUM, LIVING_GBA , ASSESSMENT_SUBNBHD, CENSUS_TRACT, CENSUS_BLOCK, GIS_LAST_MOD_DTTM, STORIES, SALE_NUM, USECODE, CITY, STATE*

**Fill data Year Remodel** 

```python
df.dropna(subset=['AYB'],inplace=True)
```

```python
group_remodel= df.groupby(['EYB','AYB']).mean()['YR_RMDL']
```

```python
# def applyRemodel(x):
#     remodel_notnull = df[(df['AYB']==x['AYB']) & (df['YR_RMDL'].notnull())]['EYB']
#     if pd.notnull(x['YR_RMDL']):
#         return x['YR_RMDL']
#     else:
#         idx = np.argmin(np.abs(remodel_notnull - x['EYB']))
#         return round(group_remodel.loc[df['EYB'][idx]][x['AYB']])
```

```python
def applyRemodel(x):
    if pd.notnull(x['YR_RMDL']):
        return x['YR_RMDL']
    else:
        return round(group_remodel.loc[x['EYB']][x['AYB']])
```

```python
df['YR_RMDL'] = df[['YR_RMDL','EYB','AYB']].apply(applyRemodel,axis = 1)
```

```python
df.dropna(subset=['YR_RMDL'],inplace=True)
```

* Bray pastikan si RMDL jadiin fitur untuk predict juga (coba dlu)
* Isi Nan dengan setiap value column terakit yee bray
* Sale date mau di jadiin apa pikirkan bray 
* Bikin rencana EDA bisa diliat dari yg udh pernah dibuat
* Predict dari setiap model bisa pake regression 

**Drop Nan Value at PRICE, LONGITUDE, ASSESSMENT_NBHD, ZIPCODE, WARD, LATITUDE**

```python
df.dropna(subset=['PRICE', 'LONGITUDE', 'ASSESSMENT_NBHD', 'ZIPCODE', 'WARD', 'LATITUDE'],inplace=True)
```

```python
df = df.reset_index().drop('index',axis=1)
```

**Drop all nan value**
<br>
Nan Value at column struct it will be same as at Nan value in another column

```python
df.dropna(subset=['STRUCT','STYLE'],inplace=True)
```

```python
df.dropna(subset=['QUADRANT'],inplace=True)
```

**Change float column to integer**

```python
df['AYB'] = df['AYB'].values.astype(int)
df['YR_RMDL'] = df['YR_RMDL'].values.astype(int)
df['NUM_UNITS'] = df['NUM_UNITS'].values.astype(int)
df['KITCHENS'] = df['KITCHENS'].values.astype(int)
df['ZIPCODE'] = df['ZIPCODE'].values.astype(int)
```

## Exploratory Data Analysis

```python
# Count by Quadrant
# Count by Ward 
# Count by ASSESSMENT_NBHD
# Ward itu wilayah/daerah gitu bray di bawahnya QUADRANT
# Neighborhood daerah sekitar dari si ward 
# Kasih Penjelasan jangan lupa bray 
```

```python
sns.set_style('whitegrid')
sns.set_context("paper", font_scale=1.3)
color = sns.color_palette("Set2", 6)
fig, ax = plt.subplots(3,1,figsize=(15,10))
sns.countplot(x='QUADRANT',data=df,ax=ax[0],order=df['QUADRANT'].value_counts().index,palette=color)
sns.countplot(x='WARD',data=df,ax=ax[1],order=df['WARD'].value_counts().index,palette=color)
sns.countplot(x='ASSESSMENT_NBHD',data=df,ax=ax[2],order=df['ASSESSMENT_NBHD'].value_counts().head(10).index,palette=color)
ax[0].set_title('Property Count by Quadrant')
ax[1].set_title('Property Count by WARD')
ax[2].set_title('TOP 10 Property Count by Neighborhood')
plt.tight_layout()
plt.show()
```
